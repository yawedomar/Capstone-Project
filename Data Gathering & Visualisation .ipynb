## Data Visualisation

### 1. Which product offering is performing better

```python
#Adding the total products
df.groupby(['product'])['quantity'].sum()
```

```python
#Plotting the Bar Chart
df.groupby(['product'])['quantity'].sum().plot(kind='bar', color='darkcyan', edgecolor='black')

#Labelling the x, and y axix
plt.xlabel('Product', fontsize=14, weight = 'bold' )
plt.ylabel('Quantity', fontsize=14, weight = 'bold')

#Giving title to the bar chart
plt.title('Product Performance',fontsize=16, weight = 'bold')

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/product_performance.png', bbox_inches='tight', dpi=120, facecolor=(1,1,1))

#Visualising the bar
plt.show()
```

#### Evaluation

From the data above, we can see that in the year 2021, toothbrushes outperformed toys. 39 toothbrushes and 27 toys were bought by customers. This indicates that, in contrast to toys, toothbushes are in high demand.

### 2. Are the sales seasonal or does it remain consistent throughout the year

#### Month vs Product Quantity

```python
# ploting a barchart with line
df.groupby(['month'])['quantity'].sum().plot(kind='bar', color='cadetblue', edgecolor='black')
df.groupby(['month'])['quantity'].sum().plot(kind='line', marker='*', color='black', ms=10)

#Labelling the x, and y axix
plt.xlabel('Month', fontsize=14, weight = 'bold')
plt.ylabel('Product Quantity', fontsize=14, weight = 'bold')

#Giving title to the bar chart
plt.title('UK Product Sales',fontsize=16, weight = 'bold')

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/product_sales.png', bbox_inches='tight', facecolor=(1,1,1))

#Visualising the bar
plt.show()
```

#### Evaluation

From the data above, evidently compared to the rest of the year, May is an especially poor month for sales. This demonstrates that sales are certainly seasonally dependent, with the summer months exhibiting a notable underperformance. The strongest months for sales are January through March, with the spring being the busiest season. The data, however, does not show which product is the most popular over the various months.

### Determining what products are bought in each month

```python
fig = plt.figure(figsize =(14, 18))

#Plotting the bar chart
df.pivot_table(index='month', columns='product', aggfunc=np.sum ).plot(kind='bar', y='quantity',
                                                         width = 0.8, edgecolor='black')                            
#Labelling the x, and y axix
plt.xlabel('Month', fontsize=14, weight = 'bold')
plt.ylabel('Quantity', fontsize=14, weight = 'bold')

#Giving title to the bar chart
plt.title('Products bought for each month', weight = 'bold', fontsize = 16)

#Changing the legend location
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/product_month.png', bbox_inches='tight', facecolor=(1,1,1))

#Visualising the bar
plt.show()
```

#### Evaluation

Here, we can see that only the months of March and October saw purchases of toys and toothbrushes. Compared to other months, January saw the highest toy buying rates.

### 4. Provide information on user demographics (age range, region) would help them hone in on marketing spend.

### Age Demographic 

```python
#Ploting the line graph
df.groupby(['age'])['quantity'].sum().plot(kind='line', color='darkslategray', marker='o')

#Labelling the x, and y axix
plt.xlabel('Age', fontsize=14,weight = 'bold')
plt.ylabel('Quantity', fontsize=14, weight = 'bold')

#Giving title to the bar chart
plt.title('Age Demographics',fontsize=16, weight = 'bold')

#Adding grid
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/age_demographics.png', bbox_inches='tight', facecolor=(1,1,1))

#Visualising the graph
plt.show()
```

#### Evaluation

The line graph above shows the total number of products purchased by different age groups. Looking at the graph, we can determine that the age group 21 and 35 bought the most overall products. However we can't determine which product was bought more frequently for each age group.

### Determining what products are bought for each age group


```python
fig = plt.figure(figsize =(14, 18))

#Plotting the bar chart
df.pivot_table(index='age', columns='product', aggfunc=np.sum ).plot(kind='bar', y='quantity',
                                                         width = 0.8, edgecolor='black')                            
#Labelling the x, and y axix
plt.xlabel('Month', fontsize=14, weight = 'bold')
plt.ylabel('Quantity', fontsize=14, weight = 'bold')


#Labelling the x and y axis
plt.xlabel('Age', fontsize=14, weight = 'bold')
plt.ylabel('Total Products', fontsize=14, weight = 'bold')

#Giving title to the bar chart
plt.title('Age vs Product', weight = 'bold', fontsize = 16)

#Changing the legend location
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/agedemongraphic_product.png', bbox_inches='tight', facecolor=(1,1,1))

#Visualising the bar
plt.show()
```

#### Evaluation

This barchart gives us a breakdown of what products were bough by each age group.
- Customers between the ages of 21 and 24 only bought toothbrushes.
- Customers between the ages of 31 and 35 only bought toys.
- Customers between the ages of 45 to 60 only bought toothbrushes

Stakeholders may want to focus on the age group that has bought the most products overall as well as the products they continue to buy in order to continue improving their overall performance.

### Regional Demographic 

```python
#Adding up the total quantity for each region
df.groupby(['uk_region'])['quantity'].sum()
```

```python
fig = plt.figure(figsize =(10, 7))

#Labelling pie chart
labels= ['UK South East', 'UK North', 'UK North East']

#Defining ratio gap
explode = (0.05, 0.05, 0.05)

#Defining pie chart colors
colors = ['darkcyan', 'mediumturquoise', 'slateblue']

#Plotting the pie chart
df.groupby(['uk_region'])['quantity'].sum().plot(kind='pie', ylabel='', 
                                    autopct='%1.0f%%', colors=colors,
                                   explode=explode,labels=None, shadow=True)

#Adding legends
plt.legend(bbox_to_anchor=(1,1), labels=labels)

#Title for the piechart
plt.title('UK Region Demographic', fontsize=16, weight = 'bold')

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/regional_demographic.png', bbox_inches='tight', facecolor=(1,1,1))

#Show plot
plt.show()
```

#### Evaluation

Toys and toothbrush purchases were primarily made by customers from the regions depicted in this pie chart. The data above demonstrates that  South East is where the majority of customers reside.The data however does not reveal which product is most popular across the different geographies. 

### Determining what products are bought in each region

```python
#Plotting the bar chart
df=df.groupby(['product','uk_region']).size()
df=df.unstack()
df.plot(kind='bar', edgecolor='black')

#Labelling the x axis
plt.xlabel('Product', fontsize=14, weight = 'bold')
plt.ylabel('Total Product', fontsize=14, weight = 'bold')


#Changing the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

#Giving title to the bar chart
plt.title('Region vs Product',fontsize=16, weight = 'bold')

#saving the graph
plt.savefig(r'C:/Users/YasminaAwedOmar/Documents/region_product.png', bbox_inches='tight', facecolor=(1,1,1))
```

#### Evaluation

We can observe that the only region that purchased toys was the South East, UK; the other region primarily purchased toothbrushes. The UK, North East region bought the most toothbrushes but no toys. The stakeholders should concentrate more of their marketing efforts on the North East region when it comes to toothbrush product, and on the South East when it comes to toys.
